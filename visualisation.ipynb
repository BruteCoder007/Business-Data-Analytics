```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
zerodha = pd.read_csv('TELIA1.HE.csv')
lycaa = pd.read_csv('TSCO.L.csv')
tesla = pd.read_csv('TSLA.csv')
```

```python

#Zerodha 
# Create a line graph
plt.plot(zerodha['Date'], zerodha['Close'])

# Add labels and a title
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Zerodha Stock Value')

# Display the graph
plt.show()
```
